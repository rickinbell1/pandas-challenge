```python
# Dependencies and Setup
import pandas as pd
# File to load(Remember to Change These)
file_to_load ="Resources/purchase_data.csv"
# Read Purchasing file and store into Pandas data Frame
purchase_data = pd.read_csv(file_to_load)


```

```python
# PLayer Count

TotalPlayers =purchase_data["SN"].nunique()
TotalPlayers



```

```python
# Purchasing Analysis(Total)

uniqueItems = purchase_data["Item ID"].nunique()
purchasePrice = purchase_data["Price"].mean()
tnumPurchase = purchase_data["Purchase ID"].count()
tRevenue = purchase_data["Price"].sum()
sumData_df = pd.DataFrame({"Number of Unique Items":[uniqueItems],"Average Price":[purchasePrice],"Number of Purchases":[tnumPurchase],"Total Revenue":[tRevenue] })
sumData_df["Average Price"] = sumData_df["Average Price"].map("${:.2f}".format)
sumData_df["Total Revenue"] = sumData_df["Total Revenue"].map("${:,.2f}".format)
sumData_df
```

```python
# Gender Demographics 

genderDemo = purchase_data.groupby("Gender")
totalCount =genderDemo["SN"].nunique()
totalSum = genderDemo["SN"].nunique().sum()
percent = round((totalCount/totalSum)* 100, 2)
demoGraph = pd.DataFrame({"Total Count":totalCount,"Percentage of Players":percent})
demoGraph["Percentage of Players"] = demoGraph["Percentage of Players"].map("{}%".format)
demoGraph

```

```python
# Purchasing Analysis(Gender)

purchaseAnalysis = purchase_data.groupby("Gender")
purchaseCount = purchaseAnalysis["Gender"].count()
avgPurPrice = purchaseAnalysis["Price"].mean()
totalpurPrice = purchaseAnalysis["Price"].sum()
avgTotalperperson = totalpurPrice /demoGraph["Total Count"]
summaryPcs = pd.DataFrame({"Purchase Count":purchaseCount,"Average Purchase Price":avgPurPrice,"Total Purchase Value":totalpurPrice,"Avg Total Purchase per Person":avgTotalperperson})
summaryPcs["Average Purchase Price"] =summaryPcs["Average Purchase Price"].map("${:.2f}".format) 
summaryPcs["Total Purchase Value"] =summaryPcs["Total Purchase Value"].map("${:.2f}".format) 
summaryPcs["Avg Total Purchase per Person"]=summaryPcs["Avg Total Purchase per Person"].map("${:.2f}".format)
summaryPcs
```

```python
# Age Demographics 

age_bins = [0, 9, 14, 19, 24, 29, 34,39, 50]
age_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
purchase_data["Age Labels"] = pd.cut(purchase_data["Age"], bins = age_bins, labels=age_labels, right = True)
agedemo = purchase_data.groupby("Age Labels")
totalcount = agedemo["SN"].nunique()
totalsum = agedemo["SN"].nunique().sum()
percentPlayer = round((totalcount/totalsum)*100,2)
ageDemo = pd.DataFrame({"Total Count":totalcount,"Percentage of Players":percentPlayer})
ageDemo["Percentage of Players"] =ageDemo["Percentage of Players"].map("{}%".format)
ageDemo

```

```python
# Purchasing Analysis(Age)

age_Pbins = [0, 9, 14, 19, 24, 29, 34,39, 50]
age_Plabels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], bins = age_Pbins, labels=age_Plabels, right = True)
agedemo1 = purchase_data.groupby("Age Ranges")
purchaseCount = agedemo1["Gender"].count()
avgPurchase_price = round(agedemo1["Price"].mean(),2)
totalPurchase_price = round(agedemo1["Price"].sum(),2)
totalCount = agedemo1["SN"].nunique()
avgTotal = round(totalPurchase_price/totalCount, 2)
purchasingAnalysis_age = pd.DataFrame({"Purchase Count":purchaseCount,"Average Purchase Price":avgPurchase_price,"Total Purchase Value":totalPurchase_price,"Avg Total Purchase per Person":avgTotal})
purchasingAnalysis_age["Average Purchase Price"] = purchasingAnalysis_age["Average Purchase Price"].map("${}".format)
purchasingAnalysis_age["Avg Total Purchase per Person"]=purchasingAnalysis_age["Avg Total Purchase per Person"].map("${}".format)
purchasingAnalysis_age["Total Purchase Value"] =purchasingAnalysis_age["Total Purchase Value"].map("${:,}".format)  
purchasingAnalysis_age




```

```python
# Top Spenders

spenders = purchase_data.groupby("SN")
purchCount =spenders["Purchase ID"].count()
purchPrice = round(spenders["Price"].mean(),2)
totalValue = round(spenders["Price"].sum(),2)
topSpenders = pd.DataFrame({"Purchase Count":purchCount,"Average Purchase Price":purchPrice,"Total Purchase Value":totalValue})
topSpenders["Average Purchase Price"]=topSpenders["Average Purchase Price"].map("${}".format)
topSpenders["Total Purchase Value"] =topSpenders["Total Purchase Value"] .map("${}".format)
topSpenders.sort_values("Total Purchase Value", ascending=False).nlargest(5, 'Purchase Count').head()
```

```python
# Most Popular Items

popularItems = purchase_data.groupby(["Item ID", "Item Name"])
purchaseCount = popularItems["Item ID"].count()
totalPValue =popularItems["Price"].sum()
itemPrice = totalPValue/purchaseCount
mPopularI = pd.DataFrame({"Purchase Count":purchaseCount,"Item Price":itemPrice,"Total Purchase Value":totalPValue})
mPopularI.sort_values("Purchase Count",ascending =False)
mPopularI["Item Price"] = mPopularI["Item Price"].map("${:.2f}".format)
mPopularI["Total Purchase Value"] = mPopularI["Total Purchase Value"].map("${:.2f}".format)
mPopularI.nlargest(5,'Purchase Count')
```

```python
# Most Profitable Items

msProfitableI = mPopularI.sort_values("Total Purchase Value", ascending = False)
msProfitableI.nlargest(5, 'Purchase Count')
```
